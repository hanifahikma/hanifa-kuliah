```python
# Membuat array 3 dimensi
# perkalian matriks array

import numpy as np

A = np.array([15, 9, 5])
B = np.array([64, 68, 54])
C = np.array([12, 14, 16])

kali = A * B * C
print(kali)
```

```python
# Pengurangan matrix

import numpy as np 

A = np.array([15, 9, 5])
B = np.array([64, 68, 54])
C = np.array([12, 14, 16])

pengurangan = A - B - C
print(pengurangan)
```

```python
# Transpose matrix

import numpy as np

A = np.array([[15, 9, 5], [64, 68, 54], [12, 14, 16]])

transpose_A = np.transpose(A)
print(transpose_A)
```

```python
# Nilai invers

import numpy as np

A = np.array([[15, 9, 5], [64, 68, 54], [12, 14, 16]])

determinant_A = np.linalg.det(A)
print("Determinan dari matriks A:", determinant_A)

inverse_A = np.linalg.inv(A)
print("\nInvers matriks A:")
print(inverse_A)
```

```python
# Nilai determinan

import numpy as np

A = np.array([[15, 9, 5], [64,68,54], [12, 14, 16]])

determinant_A = np.linalg.det(A)
print("Determinan dari matriks A:", determinant_A)
```
